## Embedding Model

````python
import pandas
# è¼‰å…¥è³‡æ–™é›†
df = pandas.read_csv('https://raw.githubusercontent.com/ywchiu/hp_ai_advanced_training/refs/heads/main/data/hp_reviews.csv')
# é¡¯ç¤ºè³‡æ–™é›†å‰å¹¾åˆ—
df.head()
````

````python
from tqdm.auto import tqdm
from openai import OpenAI
import json

client = OpenAI(api_key = API_KEY)
# ç‚º apply å‡½å¼åŠ å…¥é€²åº¦æ¢
tqdm.pandas()

def get_embedding_small(text):
  """ä½¿ç”¨ OpenAI çš„ text-embedding-3-small æ¨¡å‹ç”¢ç”Ÿæ–‡å­—åµŒå…¥ã€‚"""
  try:
    response = client.embeddings.create(
        model="text-embedding-3-small",
        input=text,
    )
    return response.data[0].embedding
  except Exception as e:
    print(f"ç”ŸæˆåµŒå…¥å‘é‡æ™‚ç™¼ç”ŸéŒ¯èª¤ï¼š{e}")
    return None # ç™¼ç”ŸéŒ¯èª¤æ™‚å›å‚³ None

# ä½¿ç”¨é€²åº¦æ¢å°‡ get_embedding å‡½å¼æ‡‰ç”¨åˆ° 'å…§å®¹' æ¬„ä½
df['åµŒå…¥å‘é‡'] = df['å…§å®¹'].progress_apply(get_embedding_small)

# é¡¯ç¤ºå¸¶æœ‰æ–°çš„ 'åµŒå…¥å‘é‡' æ¬„ä½çš„è³‡æ–™æ¡†å‰å¹¾åˆ—
display(df.head())
````

````python
from tqdm.auto import tqdm
from openai import OpenAI
import json

client = OpenAI(api_key = API_KEY)
# ç‚º apply å‡½å¼åŠ å…¥é€²åº¦æ¢
tqdm.pandas()

def get_embedding_large(text):
  """ä½¿ç”¨ OpenAI çš„ text-embedding-3-large æ¨¡å‹ç”¢ç”Ÿæ–‡å­—åµŒå…¥ã€‚"""
  try:
    response = client.embeddings.create(
        model="text-embedding-3-large",
        input=text,
    )
    return response.data[0].embedding
  except Exception as e:
    print(f"ç”ŸæˆåµŒå…¥å‘é‡æ™‚ç™¼ç”ŸéŒ¯èª¤ï¼š{e}")
    return None # ç™¼ç”ŸéŒ¯èª¤æ™‚å›å‚³ None

# ä½¿ç”¨é€²åº¦æ¢å°‡ get_embedding å‡½å¼æ‡‰ç”¨åˆ° 'å…§å®¹' æ¬„ä½
df['åµŒå…¥å‘é‡'] = df['å…§å®¹'].progress_apply(get_embedding_large)

# é¡¯ç¤ºå¸¶æœ‰æ–°çš„ 'åµŒå…¥å‘é‡' æ¬„ä½çš„è³‡æ–™æ¡†å‰å¹¾åˆ—
display(df.head())
````

````python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import numpy as np

# æå–ç‰¹å¾µè³‡æ–™ (X) å’Œç›®æ¨™æ¨™ç±¤ (y)
X = np.array(df['åµŒå…¥å‘é‡'].tolist())
y = df['æƒ…ç·’']

# å°‡æ–‡å­—æ¨™ç±¤è½‰æ›ç‚ºæ•¸å€¼
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# åˆ†å‰²è³‡æ–™é›†
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.25, random_state=42, stratify=y_encoded)

print("ç‰¹å¾µè³‡æ–™ (X) å½¢ç‹€:", X.shape)
print("ç›®æ¨™æ¨™ç±¤ (y) å½¢ç‹€:", y_encoded.shape)
print("è¨“ç·´é›†ç‰¹å¾µè³‡æ–™ (X_train) å½¢ç‹€:", X_train.shape)
print("æ¸¬è©¦é›†ç‰¹å¾µè³‡æ–™ (X_test) å½¢ç‹€:", X_test.shape)
print("è¨“ç·´é›†ç›®æ¨™æ¨™ç±¤ (y_train) å½¢ç‹€:", y_train.shape)
print("æ¸¬è©¦é›†ç›®æ¨™æ¨™ç±¤ (y_test) å½¢ç‹€:", y_test.shape)
````

````python
from sklearn.svm import SVC

# Create an SVC classifier instance
svm_classifier = SVC(kernel = 'rbf')

# Train the model using the training data
svm_classifier.fit(X_train, y_train)

print("SVM classifier trained successfully.")
````

````python
from sklearn.metrics import classification_report

# ä½¿ç”¨è¨“ç·´å¥½çš„ SVM åˆ†é¡å™¨åœ¨æ¸¬è©¦é›†ä¸Šé€²è¡Œé æ¸¬
y_pred = svm_classifier.predict(X_test)

# ä½¿ç”¨ classification_report å‡½å¼è©•ä¼°æ¨¡å‹æ•ˆèƒ½
# ä½¿ç”¨ label_encoder.classes_ ä¾†é¡¯ç¤ºæ­£ç¢ºçš„é¡åˆ¥åç¨±
report = classification_report(y_test, y_pred, target_names=label_encoder.classes_)

# å°å‡ºåˆ†é¡å ±å‘Š
print(report)
````
